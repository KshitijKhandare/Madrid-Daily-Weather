# **Suppose you're planning a vacation to Madrid and hoping for the warmest temperature. Which month might you plan to travel?**

```python
# Assuming 'CET' is the date column
df['CET'] = pd.to_datetime(df['CET'])
# Extracting the month from the date
df['Month'] = df['CET'].dt.month

# Grouping by month and calculating the average maximum temperature
average_max_temp = df.groupby('Month')['Max TemperatureC'].mean()

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(average_max_temp, marker='o', linestyle='-', color='b')
plt.title('Average Maximum Temperature in Madrid by Month')
plt.xlabel('Month')
plt.ylabel('Average Max Temperature (°C)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
average_max_temp.sort_values(ascending=True)
```

# **On which date in the sample did Madrid see the fastest wind gust? What was the weather like on that day?**

```python
# Identify the date with the fastest wind gust
max_gust_date = df.loc[df[' Max Gust SpeedKm/h'].idxmax(), 'CET']
max_gust_speed = df[' Max Gust SpeedKm/h'].max()

# Display the result
print(f"The fastest wind gust in Madrid occurred on {max_gust_date} with a speed of {max_gust_speed} km/h.")
```

```python
# Identify the date with the fastest wind gust
max_gust_date = df.loc[df[' Max Gust SpeedKm/h'].idxmax(), 'CET']
max_gust_speed = df[' Max Gust SpeedKm/h'].max()

# Plotting the bar chart
plt.bar(max_gust_date, max_gust_speed, color='purple')
plt.xlabel('Date')
plt.ylabel('Max Gust Speed (km/h)')
plt.title('Fastest Wind Gust in Madrid')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the chart
plt.show()
```

# **How does the average visibility (Km) compare for clear days vs. foggy days?**

```python
# Group the data based on the 'Events' column
grouped_data = df.groupby(' Events')[' Mean VisibilityKm'].mean().sort_values(ascending = False).round(2)
grouped_data
```

```python
grouped_data.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(grouped_data):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('Mean VisibilityKm')
plt.title('Visibility vs clear day')
```
